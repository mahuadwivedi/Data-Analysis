##### Child marriage refers to a formal or informal union where one or both parties involved are under the age of 18. It is a violation of human rights and is considered a harmful practice. This practice involves marrying children before they are physically, mentally, and emotionally mature, often against their will and without their full and free consent.

```python
datafemale_year=data.loc[data.groupby('Reference Year Female')['Married by 18 Female'].idxmax()]
datafemale_year
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation='vertical')
sns.barplot(x='Countries and areas',y='Married by 18 Female',data=datafemale_year)
```

```python
datamale_year=data.loc[data.groupby('Reference Year Male')['Married by 18 Male'].idxmax()]
datamale_year
```

```python
plt.figure(figsize=(15,7))
plt.xticks(rotation='vertical')
sns.barplot(x='Countries and areas',y='Married by 18 Male',data=datamale_year)
```

```python
frame=[datafemale_year,datamale_year]
marriage18=pd.concat(frame)
marriage18
```

```python
marriage18.shape
```

```python
marriage18=marriage18.drop_duplicates()
marriage18
```

```python
marriage18.shape
```

```python
plt.figure(figsize=(15,7))
plt.bar(marriage18['Countries and areas'],marriage18['Married by 18 Female'],0.4,label = 'Female') 
plt.bar(marriage18['Countries and areas'],marriage18['Married by 18 Male'],0.4,label = 'Male')

plt.xticks(rotation='vertical')
plt.xlabel("Countries") 
plt.ylabel("% of marriage") 
plt.title("% of marriage by 18 among Male and Female") 
plt.legend() 
plt.show() 
```

### Countries with Child Marriage above World Average

```python
average_male=data['Married by 18 Male'].mean()
round(average_male,2)
```

```python
average_female=data['Married by 18 Female'].mean()
round(average_female,2)
```

```python
dataset=data[(data['Married by 18 Female']>average_female)&(data['Married by 18 Male']>average_male)]
dataset
```

```python
plt.figure(figsize=(15,7))
plt.bar(dataset['Countries and areas'],dataset['Married by 18 Female'],0.4,label = 'Female') 
plt.bar(dataset['Countries and areas'],dataset['Married by 18 Male'],0.4,label = 'Male')
plt.axhline(average_female, color='RED',label="Female World Average")
plt.axhline(average_male, color='GREEN',label='Male World Average')

plt.xticks(rotation='vertical')
plt.xlabel("Countries") 
plt.ylabel("% of marriage") 
plt.title("Countries with Child Marriage above World Average") 
plt.legend() 
plt.show() 
```

### Observation:
* Highest % of Female child marriage i.e. marriage by age of 18 years is observed in Niger.
* Highest % of Male child marriage i.e. marriage by 18 years of age is observed in Belize.
* Overall child marriage % is reportedly highest in the Country of Niger.
* World Average for Female Child Marriage is 30.41%, which is quite high.
* World Average for Male Child Marriage is 4.7%, which is low, but still not acceptable.

### Learning Outcomes:
